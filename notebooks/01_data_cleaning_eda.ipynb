### Imports & Setup

```python
# Basic Libraries
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set(style="whitegrid")
pd.set_option("display.max_columns", None)
```

```python
# Load dataset
df = pd.read_csv("../data/raw/application_train.csv")

# Basic shape
print("Shape:", df.shape)

# Preview
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop_duplicates()
```

```python
# Check for missing values
missing = df.isnull().mean() * 100
missing = missing.sort_values(ascending=False)

missing.head(20)
```

```python
# Drop columns with >50% missing (except EXT_SOURCE_1,2,3)
cols_to_drop = missing[(missing > 50) & (~missing.index.isin(['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']))].index
df = df.drop(columns=cols_to_drop)

print("Remaining columns:", df.shape[1])
```

```python
# DAYS_EMPLOYED anomaly (365243 = missing)
df['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)
```

```python
df['TARGET'].value_counts()
```

```python
# Target distribution
sns.countplot(x='TARGET', data=df)
plt.title("Target Distribution")
plt.show()
```

```python
# Boxplot for income vs target
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[num_cols].hist(figsize=(20, 15), bins=30)
plt.tight_layout()
plt.show()
```

```python
# Analyze top 5 categorical features
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols[:5]:  # limit to avoid clutter
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col)
    plt.xticks(rotation=45)
    plt.title(col)
    plt.show()
```

```python
# Boxplot for income vs target
sns.boxplot(x='TARGET', y='AMT_INCOME_TOTAL', data=df)
plt.title("Income vs Default")
plt.show()
```

```python
# Boxplot for credit amount vs target
sns.boxplot(x='TARGET', y='AMT_CREDIT', data=df)
plt.title("Credit vs Default")
plt.show()
```

```python
# Check if EXT_SOURCE columns exist before plotting
ext_cols = [col for col in ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3'] if col in df.columns]

for col in ext_cols:
    sns.boxplot(x='TARGET', y=col, data=df)
    plt.title(f"{col} vs Default")
    plt.show()
```

```python
# Correlation heatmap
num_df = df.select_dtypes(include=['int64', 'float64'])

corr = num_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Focus on top correlated features with TARGET
top_corr = num_df.corr()['TARGET'].abs().sort_values(ascending=False).head(15).index

sns.heatmap(num_df[top_corr].corr(), cmap='coolwarm', center=0)
```

```python
target_corr = corr['TARGET'].sort_values(ascending=False)

target_corr.head(15)
```

```python
target_corr.tail(15)
```

### Insights:

Customers with low EXT_SOURCE scores have higher default risk.

Higher credit-to-income ratio may indicate higher default probability.

Income alone is not a strong predictor.

Dataset is highly imbalanced, requiring careful evaluation.

Some features contain significant missing values, requiring selection.
